```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
```

```python
(x_train,y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
x_train.shape
```

```python
y_train
```

```python
x_train[1]
```

```python
import matplotlib.pyplot as plt
plt.imshow(x_train[0])
```

```python
x_train = x_train/255
x_test = x_test/255
```

```python
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D, BatchNormalization

model = Sequential()
model.add(Conv2D(64,(3,3),activation='relu',input_shape=(28,28,1)))
model.add(BatchNormalization())
model.add(MaxPooling2D((2,2)))
model.add(Conv2D(32,(3,3),activation='relu',padding='same',stride=(1,1)))
model.add(BatchNormalization())
model.add(MaxPooling2D((2,2)))
model.add(Flatten())
model.add(Dense(128,activation='relu'))
model.add(Dense(64,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(10,activation='softmax'))
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
history = model.fit(x_train, y_train, epochs=10, validation_split=0.2)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
y_pred = y_pred.argmax(axis=1)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```

```python
import pandas as pd

train_df = pd.read_csv('sign_mnist_train.csv')
test_df = pd.read_csv('sign_mnist_valid.csv')
```

```python
y_train=train_df['label']
y_test=test_df['label']

x_train = train_df.drop('label',axis=1)
x_test = test_df.drop('label', axis=1)
```

```python
y_train = keras.utils.to_categorical(y_train, num_classes=24)
y_test = keras.utils.to_categorical(y_test, num_classes=24)
```

```python
x_train.shape
```

```python
x_train = x_train.values.reshape(-1,28,28,1)
x_test = x_test.values.reshape(-1,28,28,1)
```

```python
plt.imshow(x_train[1])
```

```python
from tensorflow.keras.layers import Dense, Flatten, Conv2D, MaxPooling2D, Dropout, BatchNormalization

model = Sequential()
model.add(Conv2D(64,(3,3),activation='relu',input_shape=(28,28,1)))
model.add(BatchNormalization())
model.add(MaxPooling2D((2,2)))
model.add(Conv2D(32,(3,3),activation='relu',padding='same',strides=(1,1)))
model.add(Dropout(0.2))
model.add(BatchNormalization())
model.add(MaxPooling2D((2,2)))
model.add(Flatten())
model.add(Dense(128,activation='relu'))
model.add(Dropout(0.3))
model.add(Dense(64,activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(24,activation='softmax'))
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test))
```
