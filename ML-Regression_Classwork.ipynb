## Multiple Linear Regression 
## (y = m1x1+m2x2+m3x3+...+c)

```python
import pandas as pd
import seaborn as sns

df = pd.read_csv('CAR DETAILS FROM CAR DEKHO.csv')
df.head()
```

```python
df.drop('name',axis=1,inplace=True)
```

```python
#One-hot encoding
df = pd.get_dummies(df,drop_first=True)
```

```python
df.shape
```

```python
x = df.drop('selling_price',axis=1)
y = df['selling_price']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,
                                                   random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
print(lr.coef_)
```

```python
print(lr.intercept_)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error 
print(mean_absolute_error(y_test,y_pred)) # sum(|Ytest-Ypred|)/n
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test,y_pred)) # sum((ytest-ypred)^2)/n
```

```python
from sklearn.metrics import r2_score
print(r2_score(y_test,y_pred)) # 1 - (sum(ytest-ypred)/sum(ytest-ymean))
```

```python
x1,x2,x3 --> y = m1x1^2 + m2x2^2 + m3x3^2 + m4x1x2 + m5x2x3 + m6x1x3 + 
m7x1 + m8x2 + m9x3 +c

n = no. of features (x)
d = no. of degree
poly features = (n+d)!/(n!*d!)
```

```python
df = pd.read_csv('insurance.csv')
from sklearn.pipeline import Pipeline
df = df.loc[df['bmi']!='-']
df['bmi'] = df['bmi'].astype('float')
df = pd.get_dummies(df,drop_first=True)
x = df.drop('charges',axis=1)
y = df['charges']
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=10)
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train_poly,y_train)
y_pred = lr.predict(x_test_poly)
print(lr.coef_)
print(lr.intercept_)
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
print(mean_squared_error(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
pred = {}
mse_data = []
x_new = pd.DataFrame({'Wind speed (m/s)':[5,10,15,20]})
for i in range(3,7):
    poly = PolynomialFeatures(degree=i)
    x_train_poly = poly.fit_transform(x_train)
    x_test_poly = poly.transform(x_test)
    x_new_poly = poly.transform(x_new)
    model = LinearRegression()
    model.fit(x_train_poly,y_train)
    print(f'coeffecient for degree {i} is',model.coef_)
    print(f'intercept for degree {i} is',model.intercept_)
    y_pred = model.predict(x_test_poly)
    y_new = model.predict(x_new_poly)
    print(f'Predicted values for windspeed 5,10,15,20 for degree {i} is',y_new)
    pred[i] = y_pred
    mse = mean_squared_error(y_test,y_pred)
    mse_data.append(mse)
    r2 = r2_score(y_test,y_pred)
    print(f'MSE for degree {i} is', mse)
    print(f'r2 score for degree {i} is',r2)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x_test,y_test)
for i in range(3,7):
    plt.scatter(x_test,pred[i])
plt.show()
```

```python
deg = [3,4,5,6]
plt.plot(deg,mse_data)
plt.show()
```

```python
df = pd.read_csv('Marksheet.csv')
df.head()
```

```python
df.info()
```

```python
df['TEST-3 MARKS'] = df['TEST-3 MARKS'].replace('AB',0)
df['TEST-4 MARKS'] = df['TEST-4 MARKS'].replace('AB',0)
```

```python
df['TEST-3 MARKS'] = df['TEST-3 MARKS'].astype('float')
df['TEST-4 MARKS'] = df['TEST-4 MARKS'].astype('float')
```

```python
df['Total Marks'] = df['TEST-1 MARKS'] + df['TEST-2 MARKS'] + df['TEST-3 MARKS'] + df['TEST-4 MARKS']
df.head()
```

```python
df['Grade'] = np.where(df['Total Marks']>85,'A',
                      np.where(df['Total Marks']>75,'B',
                              np.where(df['Total Marks']>65,'C',
                                      np.where(df['Total Marks']>50,'D',
                                              np.where(df['Total Marks']>35,'E','F')))))
df.head()
```

```python
df.groupby('Div')['Total Marks'].mean()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

x = df[['TEST-1 MARKS']]
y = df['Total Marks']
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
print(model.coef_)
print(model.intercept_)
print(mean_squared_error(y_test,y_pred))
print(r2_score(y_test,y_pred))
```

```python
r2 = []
for i in range(0,51):
    x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,
                                                        random_state=i)
    model = LinearRegression()
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    r2.append(r2_score(y_test,y_pred))
plt.plot(r2)
print(max(r2))
print(r2.index(max(r2)))
```
